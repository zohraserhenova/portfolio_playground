# Playing with the dataset: https://www.kaggle.com/danielgrijalvas/movies

```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read in the data

df = pd.read_csv("C:/Users/ZSerkhenova/Desktop/pandas/movies.csv")
```

```python
# Look into the data details

df.head()
```

```python
# Checking if there is any missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Data types for the columns herein

df.dtypes
```

```python
# Change columns data types
# First, need to fill nAn/NA in order to avoid mistakes when applying "astype"
df = df.fillna(0)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Creating correct year column

df['correct_year'] = df['released'].astype('str').str[:11]
```

```python
df = df.drop(['correct_year'], axis = 1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# Scatter plot with Budget vs. Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs. Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Plotting Budget vs. Gross using seaborn2

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

### Looking at correlation

```python
df.corr(method='spearman') #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking into company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
